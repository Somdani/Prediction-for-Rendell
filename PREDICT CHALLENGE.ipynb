```python
train_df.drop("Valencia_pressure",axis=1,inplace=True)
```

We have removed **Valencia_pressure** from the data and replace it with a **Valencia_pressure_median**

```python
dummies = pd.get_dummies(train_df.Valencia_wind_deg)
dummies = pd.get_dummies(train_df.Seville_pressure)
train_df = train_df.drop('Valencia_wind_deg', axis=1)
train_df = train_df.drop('Seville_pressure', axis=1)
train_df = train_df.merge(dummies, left_index=True, right_index=True)
```

```python
train_df.sample()
```

```python
dummies = pd.get_dummies(test_df.Valencia_wind_deg)
dummies = pd.get_dummies(test_df.Seville_pressure)
test_df = test_df.drop('Valencia_wind_deg', axis=1)
test_df = test_df.drop('Seville_pressure', axis=1)
test_df = test_df.merge(dummies, left_index=True, right_index=True)
```

```python
test_df.sample()
```

We have now coverted our categorical variables into dummies (binary values) and created new features related to categorical values.

```python
df = pd.concat([train_df, test_df])
df.sample()
```

We have combine both train and test data. However, we note that test data did not have target varible and these may lead to this feature to have missing values.

```python
df['load_shortfall_3h'].isnull().sum()
```

We can see that the feature **load_shortfall_3h** is now having missing values that are exactly the same as the number of columns for our test data.

```python
y = df[:len(train_df)][['load_shortfall_3h']]
X = df[:len(train_df)].drop('load_shortfall_3h', axis=1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=100)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

We have standardized our independant variables of our training data to make sure that the data is consistent, that is all values has the same format since its not easy to compare the values.

```python
from sklearn import linear_model

reg_model = linear_model.LinearRegression()
reg_model.fit(X_train, y_train)
```

We have fitted our regression model using our independat and dependant training samples.

```python
predictions = reg_model.predict(X_test)
```

We then use our indepedant variables for test sample to perform our predictions.

```python
from sklearn.metrics import mean_squared_error, r2_score

def rmse(y_test, y_pred):
    return np.sqrt(mean_squared_error(y_test, y_pred))
```

```python
import sklearn.metrics as sm
print("Mean absolute error =", round(sm.mean_absolute_error(y_test, predictions), 2)) 
print("Root mean square error =", round(rmse(y_test, predictions), 2))  
print("R2 score =", round(sm.r2_score(y_test, predictions), 2))
```

Above we have calculated mean absolute error, root mean square error and R-squared for our regression model.

```python
x_train = df[:len(train_df)].drop('load_shortfall_3h',axis=1)
x_test = df[len(train_df):].drop('load_shortfall_3h',axis=1)
```

```python
reg_model.fit(x_train, y)
```

We are now using our test samples which is 25% of our original data to validate the model and predict the target variable which we did not have at first in our test data.

```python
predictions = reg_model.predict(x_test)
```
